```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set(rc = {'figure.figsize':(12,9)})
sns.set(font_scale = 2)
sns.set_style("white")
```

# Import User level data 

```python
df_user = pd.read_csv('/Users/ziyufan/Documents/Fall2/Ml/Project/data/train_user.csv', index_col=0)
```

## Website visitors behavior

### Mean Page views 

```python
fig = sns.distplot( df_user[df_user['is_transaction']==0]['pageviews_mean'], color="tomato", label="no_trans")
fig = sns.distplot( df_user[df_user['is_transaction']==1]['pageviews_mean'] , color="skyblue", label="trans")
fig.set(xlim=(0,80))
plt.legend()
plt.title('Users Mean Page views')
fig.figure.savefig("users_page_views_mean.png")
```

### Mean hits

```python
fig = sns.distplot( df_user[df_user['is_transaction']==0]['hits_mean'], color="tomato", label="no_trans")
fig = sns.distplot( df_user[df_user['is_transaction']==1]['hits_mean'] , color="skyblue", label="trans")
fig.set(xlim=(0,120))
plt.legend()
plt.title('Users Mean Hits')
fig.figure.savefig("users_hits_mean.png")
```

## sessions

### Last session

```python
fig = sns.distplot( df_user[df_user['is_transaction']==0]['last_session_1_adjmean'], color="tomato", label="no_trans")
fig = sns.distplot( df_user[df_user['is_transaction']==1]['last_session_1_adjmean'] , color="skyblue", label="trans")
fig.set(ylim=(0,1), xlim=(0,20))
plt.legend()
plt.title('Users Mean Last Session')
fig.figure.savefig("users_last_session_mean.png")
```

CDF Plot

```python
fig = sns.distplot(df_user[df_user['is_transaction']==0]['last_session_1_adjmean'], color="tomato", label="no_trans",
                   hist_kws=dict(cumulative=True), kde_kws=dict(cumulative=True))
fig = sns.distplot(df_user[df_user['is_transaction']==1]['last_session_1_adjmean'] , color="steelblue", label="trans",
                   hist_kws=dict(cumulative=True), kde_kws=dict(cumulative=True))
plt.legend()
plt.title('Users Mean Last Session')
fig.figure.savefig("users_last_session_1_adjmean_cdf.png")  
```

### Next session

```python
fig = sns.distplot( df_user[df_user['is_transaction']==0]['next_session_1_adjmean'], color="tomato", label="no_trans")
fig = sns.distplot( df_user[df_user['is_transaction']==1]['next_session_1_adjmean'] , color="skyblue", label="trans")
fig.set(ylim = (0,1), xlim = (0,20))
plt.legend()
plt.title('Users Mean Next Session')
fig.figure.savefig("users_next_session_mean.png")
```
