```python
import pandas as pd
import json
```

```python
train_set = pd.read_csv('train.csv',low_memory=False)
```

```python
def geo_network_split(df):
    df_1 = train_set.geoNetwork.map(lambda row: json.loads(row))
    df_2 = pd.DataFrame(df_1.values.tolist(), index=df.index)
    df_new = pd.concat([df, df_2], axis=1).drop('geoNetwork', axis=1)
    return df_new
```

Split geoNetwork into city/cityid/continent/country/latitude/longitude/metro/networkDomain/networkLocation/region/subContinent

```python
train_set = geo_network_split(train_set)
```

```python
geo_network = train_set[['city','cityId','continent','country','latitude','longitude','metro','networkDomain','networkLocation','region','subContinent']]
geo_network
```
